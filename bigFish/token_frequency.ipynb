```python
import nltk
import re
```

```python
def sort_dict(dl):
    greatest = 0
    clean = []
    while(len(dl) != 0):
        temp = dl[0]
        greatest = 0
        for i in range(len(dl)):
            if dl[i][1] > temp[1]:
                temp = dl[i]
                greatest = i
        clean.append(temp)
        dl.pop(greatest)
    return clean    
```

```python
f = open('fraud.txt', 'r')
```

```python
temp = []

for item in f:
    temp.append(item)
    
tempstr = ""

for item in temp: tempstr += item
```

```python
from nltk.probability import ConditionalFreqDist
from nltk.tokenize import word_tokenize
sent = tempstr
cfdist = ConditionalFreqDist()
for word in word_tokenize(sent):
    condition = len(word)
    cfdist[condition][word] += 1
```

```python
q = []


for item in x.items():
    q.append(item)
```
